```python
import matplotlib.pyplot as plt
import numpy as np

from cmap_factory import get_preset_color_pool
```

```python
chinese_color_pool = get_preset_color_pool('chinese')
```

```python
ShanLan_to_GanLanShiLv = chinese_color_pool.get_color_list(n_0='闪蓝',
                                                           n_1='橄榄石绿',
                                                           threshold=20,
                                                           same_weight=True)
ShanLan_to_GanLanShiLv
```

```python
GanLanShiLv_to_DanTengLuoZi = chinese_color_pool.get_color_list(
    n_0='橄榄石绿', n_1='淡藤萝紫', threshold=20, same_weight=True)
GanLanShiLv_to_DanTengLuoZi
```

```python
ShanLan_to_GanLanShiLv_to_DanTengLuoZi = ShanLan_to_GanLanShiLv[:-1] + GanLanShiLv_to_DanTengLuoZi
ShanLan_to_GanLanShiLv_to_DanTengLuoZi
```

```python
cmap = ShanLan_to_GanLanShiLv_to_DanTengLuoZi.to_cmap(mode='linear', name='晴日藤萝')
cmap
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)

plt.imshow(Z, cmap=cmap)
```
